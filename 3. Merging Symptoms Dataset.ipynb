# First Dataset

```python
import pandas as pd

df = pd.read_csv('Datasets/Final_Train_Data.csv')
```

```python
df['prognosis'].value_counts()
```

```python
df['Unnamed: 0'].value_counts()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Replace spaces in column names with underscores
df.columns = df.columns.str.replace(" ", "_")

# Iterate over the first 400 columns and modify their values
for col in df.columns[:-1]:  # Exclude the last column ('prognosis')
    df[col] = df[col].apply(lambda x: col if x == 1 else "")
```

```python
df.head()
```

```python
df.info()
```

```python
# Replace NaN values with an empty string (if necessary)
df = df.fillna("")

# Concatenate all symptoms into a single 'Symptom' column
df['Symptom'] = df.iloc[:, :-1].apply(lambda row: " ".join(row), axis=1)
```

```python
df.head()
```

```python
df['Symptom'][0]
```

```python
df = df.drop(df.columns[:-2], axis=1)
```

# Second Dataset

```python
df2 = pd.read_csv('Datasets/dataset.csv')
df2.head()
```

```python
df2.info()
```

```python
df2['Disease'].value_counts()
```

```python
df2 = df2.fillna("")
df2['Symptom']=""
for i in range(1,18):
    df2['s']=df2["Symptom_{}".format(i)]
    df2['Symptom']=df2['Symptom']+df2['s']
```

```python
df2.head()
```

```python
for i in range(1,18):
    df2=df2.drop("Symptom_{}".format(i),axis=1)
df2=df2.drop("s",axis=1)
```

```python
df2.head()
```

```python
df2.info()
```

```python
df.columns
```

```python
df2.columns
```

```python
# Rename the 'prognosis' column in df to 'Disease'
df = df.rename(columns={'prognosis': 'Disease'})

# Concatenate the two DataFrames
combined_df = pd.concat([df, df2], axis=0, ignore_index=True)
```
