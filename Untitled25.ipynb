```python
import random

def print_board(board):
  print(f"{board[0]} | {board[1]} | {board[2]}")
  print("---------")
  print(f"{board[3]} | {board[4]} | {board[5]}")
  print("---------")
  print(f"{board[6]} | {board[7]} | {board[8]}")
  print("---------")

def check_win(board, player):
  win_conditions = [
    [0, 1, 2], [3, 4, 5], [6, 7, 8],
    [0, 3, 6], [1, 4, 7], [2, 5, 8],
    [0, 4, 8], [2, 4, 6]
  ]
  for condition in win_conditions:
    if all(board[i] == player for i in condition):
      return True
  return False

def dfs(board, player):
  if check_win(board, 'X'):
    print_board(board)
    print("X wins!")
    return
  elif check_win(board, 'O'):
    print_board(board)
    print("O wins!")
    return

  for i in range(9):
    if board[i] == ' ':
      board[i] = player
      print_board(board)
      if player == 'X':
        dfs(board, 'O')
      else:
        dfs(board, 'X')

      board[i] = ' '
      return

  print_board(board)
  print("It's a tie!")

def computer_move(board):
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            if check_win(board, 'O'):
                return
            board[i] = ' '


    for i in range(9):
        if board[i] == ' ':
            board[i] = 'X'
            if check_win(board, 'X'):
                board[i] = 'O'
                return
            board[i] = ' '


    available_moves = [i for i in range(9) if board[i] == ' ']
    move = random.choice(available_moves)
    board[move] = 'O'


board = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
print_board(board)


while True:
  player_move = int(input("Enter your move (0-8): "))
  if 0 <= player_move <= 8 and board[player_move] == ' ':
      board[player_move] = 'X'
      print_board(board)
      if check_win(board, 'X'):
          print("You win!")
          break
      elif all(cell != ' ' for cell in board):
          print("It's a tie!")
          break
      else:
          computer_move(board)
          print_board(board)
          if check_win(board, 'O'):
              print("Computer wins!")
              break
          elif all(cell != ' ' for cell in board):
              print("It's a tie!")
              break
  else:
      print("Invalid move. Try again.")
```
